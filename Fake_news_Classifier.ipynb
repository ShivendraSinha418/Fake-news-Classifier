```python
from bs4 import BeautifulSoup
import requests as rq
import nltk
import pandas as pd
import gensim
import numpy as np
```

```python
links = ["https://www.bbc.com/news","https://www.bbc.com/sport","https://www.bbc.com/business","https://www.bbc.com/innovation","https://www.bbc.com/culture","https://www.bbc.com/news/topics/c2vdnvdg6xxt","https://www.bbc.com/news/war-in-ukraine","https://www.bbc.com/news/topics/cj3ergr8209t","https://www.bbc.com/news/us-canada","https://www.bbc.com/news/uk","https://www.bbc.com/news/world/africa","https://www.bbc.com/news/world/asia","https://www.bbc.com/news/world/europe","https://www.bbc.com/news/world/latin_america","https://www.bbc.com/news/world/middle_east","https://www.ndtv.com/Latest","https://www.ndtv.com/india","https://www.livemint.com/market/stock-market-news","https://www.indiatvnews.com/entertainment/bollywood","https://sports.ndtv.com/cricket/news","https://indianexpress.com/section/technology/science","https://economictimes.indiatimes.com/topic/artificial-intelligence"]
```

```python
Latest_Real_News =[]
for link in links:
    r = rq.get(link)
    soup = BeautifulSoup(r.text,"html")
    if(soup.find_all("h2",class_="sc-4fedabc7-3 dsoipF")):
        temp_top_head = soup.find_all("h2",class_="sc-4fedabc7-3 dsoipF")
        for news in temp_top_head:
            Latest_Real_News.append(news.text)
    if(soup.find_all("p",class_="sc-f98732b0-0 iQbkqW")):
        temp_top_desc = soup.find_all("p",class_="sc-f98732b0-0 iQbkqW")
        for desc in temp_top_desc:
            Latest_Real_News.append(desc.text)
    if(soup.find_all("h2",class_="sc-4fedabc7-3 zTZri")):
        temp_insider = soup.find_all("h2",class_="sc-4fedabc7-3 zTZri")
        for news in temp_insider:
            Latest_Real_News.append(news.text)
    if(soup.find_all("p",class_="sc-b8778340-4 kYtujW")):
        temp_ins_desc = soup.find_all("p",class_="sc-b8778340-4 kYtujW")
        for desc in temp_ins_desc:
            Latest_Real_News.append(desc.text)
    if(soup.find_all("h2",class_="sc-4fedabc7-3 bvDsJq")):
        temp_bottom = soup.find_all("h2",class_="sc-4fedabc7-3 bvDsJq")
        for news in temp_bottom:
            Latest_Real_News.append(news.text)
    if(soup.find_all("p",class_="sc-ae29827d-0 cNPpME")):
        temp_bot_desc = soup.find_all("p",class_="sc-ae29827d-0 cNPpME")
        for desc in temp_bot_desc:
            Latest_Real_News.append(desc.text)
    if(soup.find_all("h2",class_="newsHdng")):
        temp_ndtv_lat = soup.find_all("h2",class_="newsHdng")
        for i in range(len(temp_ndtv_lat)):
            lat = temp_ndtv_lat[i].find("a")
            Latest_Real_News.append(lat.text)
    if(soup.find_all("p",class_="newsCont")):
        temp_nl_dec = soup.find_all("p",class_="newsCont")
        for desc in temp_nl_dec:
            Latest_Real_News.append(desc.text)
    if(soup.find_all("h2",class_="newsHdng")):
        temp_ndtv_india = soup.find_all("h2",class_="newsHdng")
        for i in range(len(temp_ndtv_india)):
            ind = temp_ndtv_india[i].find("a")
            Latest_Real_News.append(ind.text)
    if(soup.find_all("h2",class_="headline")):
        stock = soup.find_all("h2",class_="headline")
        for i in range(len(stock)):
            abt = stock[i].find("a")
            Latest_Real_News.append(abt.text)
    if(soup.find_all("h3",class_="title")):
        bollywood_desc = soup.find_all("h3",class_="title")
        for i in range(len(bollywood_desc)):
            bst_d = bollywood_desc[i].find("a")
            Latest_Real_News.append(bst_d.text)
    if(soup.find_all("p",class_="dic")):
        boll_D = soup.find_all("p",class_="dic")
        for i in boll_D:
            Latest_Real_News.append(i.text)
    if(soup.find_all("a",class_="lst-pg_ttl")):
        crik = soup.find_all("a",class_="lst-pg_ttl")
        for i in crik:
            Latest_Real_News.append(i.text)
    if(soup.find_all("p",class_="lst-pg_txt txt_tct txt_tct-three")):
        crik_d = soup.find_all("p",class_="lst-pg_txt txt_tct txt_tct-three")
        for i in crik_d:
            Latest_Real_News.append(i.text)
    if(soup.find_all("h2",class_="title")):
        rnd = soup.find_all("h2",class_="title")
        for i in range(len(rnd)):
            rvd = rnd[i].find("a")
            Latest_Real_News.append(rvd.text)
    if(soup.find_all("p",class_="wrapLines l3")):
        ai = soup.find_all("p",class_="wrapLines l3")
        for i in ai:
            Latest_Real_News.append(i.text)
```

```python
fake_data = pd.read_csv("Fake.csv")
fake_news_articles = fake_data['text']
```

```python
news_text =[]
news_label =[]
for news in Latest_Real_News:
    news_text.append(news)
for news in fake_news_articles:
    news_text.append(news)
for i in range(len(Latest_Real_News)):
    news_label.append(1)
for i in range(len(fake_news_articles)):
    news_label.append(0)
label = np.array(news_label)
```

```python
import re
from nltk import WordNetLemmatizer
from nltk.corpus import stopwords
stop = stopwords.words("english")
```

```python
pure_corpus =[]
for i in range(len(news_text)):
    refine = re.sub("[^A-Za-z]",' ',news_text[i])
    refine = refine.lower()
    pure_corpus.append(refine)
```

```python
n_data =[]
for sen in pure_corpus:
    sen_token = nltk.word_tokenize(sen)
    terms =[]
    for word in sen_token:
        if(word not in stop):
            terms.append(word)
    n_data.append(terms)    
```

```python
from gensim import downloader as api
model = api.load("word2vec-google-news-300")
```

```python
news_vec =[]
for sentence in n_data:
    features = np.zeros(300)
    for i in range(len(sentence)):
        if(sentence[i] in model.key_to_index):
            temp = model[sentence[i]]
            for j in range(300):
                features[j] = features[j] + temp[j]
    news_vec.append(features)
news_vec = np.array(news_vec)
# def avgWord2vec(doc):
#     return np.mean([model['word'] for word in doc if word in model.key_to_index],axis=0)
```

```python
X =[]
for vec in news_vec:
    X.append(vec)
X = np.array(X)
Y= label
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
rfc.score(X_test,Y_test)*100
```
